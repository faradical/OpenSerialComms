```python
import serial
import time

ser = serial.Serial("COM6", 11520, timeout=1)

while True:
    # simulate device sending data
    ser.write(b"TEMP:24.3\n")
    time.sleep(2)

    # read commands from your program
    if ser.in_waiting:
        cmd = ser.readline().decode().strip()
        print("Received:", cmd)

        if cmd == "PING":
            ser.write(b"PONG\n")
```

```python
import serial.tools.list_ports

for p in serial.tools.list_ports.comports():
    print(p.device, p.description)
```
